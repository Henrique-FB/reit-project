```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

from selenium import webdriver

import ast


```

```python
URL = "https://fiis.com.br/rendimentos/?ticker="
dr = webdriver.Chrome()
dr.implicitly_wait(3)
```

```python
dr.get(URL)
```

```python
soup = BeautifulSoup(dr.page_source, "lxml")
output = pd.DataFrame()
dr.quit()
```

```python
FII_list = soup.findAll("div",attrs={"class":"accordion-body"})
FII_name_list =[item['data-accordion-body'] for item in FII_list]
missing_FIIs = []
```

```python
FII_list[0].find("div", attrs={"data-component":"paginator"})["data-options"]
```

```python
for FII in FII_list:
    try:
        temp_data = FII.find("div", attrs={"data-component":"paginator"})
        temp = pd.DataFrame(ast.literal_eval(temp_data["data-options"])["data"]["items"][0])
        output = pd.concat([output, temp], ignore_index=True)
    except:
        missing_FIIs.append(FII['data-accordion-body'])

```
